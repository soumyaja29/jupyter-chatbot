# 🤖 Simple Rule-Based Chatbot in Jupyter
This notebook contains a basic chatbot implementation using Python and interactive widgets. It uses simple rule-based logic for conversation.

```python
# 📚 Define Simple Rule-Based Chatbot Logic
def chatbot_response(user_input):
    user_input = user_input.lower()
    
    if 'hello' in user_input or 'hi' in user_input:
        return "Hello! How can I help you today?"
    elif 'your name' in user_input:
        return "I'm a simple chatbot running in a Jupyter Notebook."
    elif 'weather' in user_input:
        return "I'm not connected to the internet, but it looks sunny inside this notebook ☀️"
    elif 'bye' in user_input or 'exit' in user_input:
        return "Goodbye! Have a great day!"
    else:
        return "I'm not sure how to respond to that. Try asking something else."
```

```python
# 💬 Interactive Chat Interface
import ipywidgets as widgets
from IPython.display import display, clear_output

chat_history = widgets.Output()
user_input = widgets.Text(placeholder='Type your message here...')

def on_submit(change):
    with chat_history:
        print(f'You: {user_input.value}')
        response = chatbot_response(user_input.value)
        print(f'Bot: {response}\n')
    user_input.value = ''

user_input.on_submit(on_submit)
display(chat_history, user_input)
```
